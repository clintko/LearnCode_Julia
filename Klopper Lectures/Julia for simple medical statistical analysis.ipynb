# Basic Operation of DataFrame

## Defining a DataFrame

```julia
# a dataframe contain 2 columns, named A and B
df = DataFrame(A = 1:10, B = 2:2:20)
```

## Selecting certain rows and columns

```julia
# selecting only rows 1 to 3 from column A
df[1:3, [:A]]
```

## Adding a column named Names containing strings

```julia
df[:Names] = ["A", "A", "A", "B", "C", "B", "A", "B", "C", "A"]
```

## Creating a new DataFrame by selecting only values of A in Names

```julia
data_A = df[df[:Names] .== "A", :]
```

## Multiple selection criteria

```julia
data_A_B = df[(df[:Names] .== "A" ) .& (df[:B] .> 4), :]
```

# Populating an empty dataframe

```julia
data = DataFrame() # creating the empty dataframe
```

```julia
# populating an array with a single string value
gender_array = ["F"] 

# Creating a for-loop to randomly add M and F string values to the gender array
for n = 1:199
    rn = randn()
    if (rn <= 0)
        append!(gender_array, ["F"])
    else
        append!(gender_array, ["M"])
    end
end

# show the result
gender_array
```

```julia
# adding gender array as a column in the data dataframe
data[:Gender] = gender_array;
```

```julia
group_array = ["A"]

# Creating a for-loop to randomly add M and F string values to the gender array
for n = 1:199
    rn = randn()
    if (rn <= 0)
        append!(group_array, ["B"])
    else
        append!(group_array, ["A"])
    end
end

# show
group_array
```

```julia
data[:Group] = group_array;
```

add more columns

```julia
data[:Age]  = rand(Normal(35, 10), 200)
data[:Days] = rand(Poisson(2), 200)
data[:Temp] = rand(Normal(38.5, 2), 200)
data[:WCC]  = rand(Normal(15, 5), 200)
data[:CRP]  = rand(Poisson(2), 200) * 2 + 100;
```

```julia
data
```

# Gadfly Plot

```julia
plot(data, 
    x = :Gender,
    #color = :Gender, 
    Geom.histogram,
    Guide.title("Gender"),
    Guide.ylabel("Number"))
```

```julia
plot(data, 
    x = "Gender",
    y = "Age", 
    Geom.boxplot,
    Guide.title("Age difference between male and female"))
```

```julia
plot(data, 
    x = "WCC", # white cell count
    color = "Gender", 
    Geom.density,
    Guide.title("WCC distribution in males and females"))
```

```julia
plot(data, 
    x = "WCC", # white cell count
    color = "Group", 
    Geom.density,
    Guide.title("WCC distribution in A and B"))
```

```julia
plot(data, 
    x = "CRP",
    Geom.density,
    Guide.title("CRP distribution for the complete dataset"))
```

```julia
plot(data, 
    x = "WCC",
    y = "CRP", 
    color = "Group",
    Geom.point,
    Guide.title("Correlation between WCC & CRP in groups A & B"))
```

```julia
plot(data, x = "Group", y = "CRP", Geom.point, Scale.x_discrete)
```
